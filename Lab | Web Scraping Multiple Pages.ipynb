#### 1. Create a Python list with the top ten FBI's Most Wanted names: url = 'https://www.fbi.gov/wanted/topten'

```python
# 2. find url and store it in a variable
url = "https://www.fbi.gov/wanted/topten"
```

```python
# 3. download html with a get request
response = requests.get(url) #gets the html and puts in response
#response.status_code
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# 4.2 check how it looks
#soup
```

```python
# 5. retrieve/extract the desired info 
    # 1. Name of the artist
    # 2. Title of the song
    
soup.select("#query-results-0f737222c5054a81a120bce207b0446a > ul > li:nth-child(1) > h3")
```

```python
soup.select("h3")[0].get_text()
```

```python
name = []

num_iter = len(soup.select("h3.title a"))

namelist = soup.select("h3.title a")
for i in range(num_iter):
    name.append(namelist[i].get_text())

print(name)
```

```python
names = pd.DataFrame({"name":name})
names.head()
```
